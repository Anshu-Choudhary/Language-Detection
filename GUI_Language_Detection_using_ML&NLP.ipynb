```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

def detect():
    lan_df=pd.read_csv('language.csv')
    x=np.array(lan_df['Text'])
    y=np.array(lan_df['language'])
    cv=CountVectorizer()
    X=cv.fit_transform(x)
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
    Model=MultinomialNB()
    Model.fit(X_train,y_train)
    user=ent.get()


    
    data=cv.transform([user]).toarray()
    output=Model.predict(data)
    lb2=Label(win,text=str(output),font=('Elephant',20,'bold'))
    lb2.place(x=200,y=305)
```

```python
from tkinter import*
win=Tk()
win.geometry("500x500")
win.title("Language Detection")
ent=StringVar()
topframe=Frame(win,width=200,height=100,bg='yellow',relief='raise',bd=5)
topframe.pack(side=TOP)
lbtop=Label(topframe,text="Language_Detection",font=('Elephant',20,'bold'),width=25,bd=4,bg='yellow')
lbtop.grid(row=0,column=0)
lb1=Label(win,text="Enter any text",font=("Elephant",20),width=23)
lb1.place(x=40,y=55)
tb1=Entry(win,textvariable=ent,font=("Elephant",15,'bold'),width=35,justify=CENTER)
tb1.place(x=0,y=100,height=100)
btn=Button(win,text="Detect",font=("Elephant",20,'bold'),command=detect)
btn.place(x=200,y=200)
win.mainloop()
```
